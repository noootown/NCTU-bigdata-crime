```python
import numpy as np
import pandas as pd
import time
```

```python
data = pd.DataFrame.from_csv("./data/CrimesAll.csv")
print("Original Length: ", len(data))
data = data.dropna(axis=0)
print("Dropna Length: ", len(data))
data.head(1)
```

```python
#cutoff outliers
data = data[data['Latitude'] > 41.644]
data = data[data['Latitude'] < 42.020]
data = data[data['Longitude'] > -87.940]
data = data[data['Longitude'] < -87.521]
print("Cut off outliers:", len(data))
```

```python
new_data = data.drop(data.columns[[0, 2, 6, 7, 8, 13, 14, 15, 17, 18, 19, 20]], axis=1)
new_data.head()
```

```python
len(new_data)
```

```python
formate = '%m/%d/%Y %I:%M:%S %p'
def time_obj(dfrow):
    obj = time.strptime(dfrow, formate)
    new_obj = time.strftime("%Y-%m",obj)
    return new_obj
```

```python
new_data['DateIdx'] = list(map(time_obj, new_data['Date']))
new_data.head()
```

```python
new_data = new_data.set_index('DateIdx')
new_data.to_csv('ModifyCrime.csv')
```
